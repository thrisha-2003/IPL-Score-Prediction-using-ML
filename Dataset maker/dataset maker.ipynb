```python
import pandas as pd

# Load the CSV files
deliveries = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')
ipl = pd.read_csv('ipl.csv')
```

```python
ipl
```

```python
merged_df = pd.merge(deliveries, matches, left_on='match_id', right_on='id', how='inner')
merged_df
```

```python
grouped_df = merged_df.groupby(['match_id', 'inning', 'over']).agg({
    'total_runs': 'sum',
    'is_wicket': 'sum'
}).reset_index()
```

```python
grouped_df
```

```python
total_runs_by_inning = merged_df.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
total_runs_by_inning.columns = ['match_id', 'inning', 'total_runs']
```

```python
total_runs_by_inning
```

```python
import pandas as pd



# Initialize a list to store results
results = []

# Group by match_id and inning
for (match_id, inning), group in grouped_df.groupby(['match_id', 'inning']):
    # Ensure there are at least 5 overs in the entire inning
    if len(group) >= 5:
        # Initialize cumulative sum for runs
        cumulative_runs = 0
        
        # Loop through each over in the entire inning
        for index, row in group.iterrows():
            over = row['over']
            runs = row['total_runs']
            
            # Add the runs of the current over to the cumulative sum
            cumulative_runs += runs
            
            # Consider overs from 5 to 20
            if 5 <= over <= 20:
                # Filter for the last 5 overs dynamically
                last_five_overs = group[(group['over'] >= over - 4) & (group['over'] <= over)]
                
                # Ensure there are exactly 5 overs in the last five overs window
                if len(last_five_overs) == 5:
                    last_5_over_sum = last_five_overs['total_runs'].sum()
                    last_5_over_wickets = last_five_overs['is_wicket'].sum()
                    
                    result = {
                        'match_id': match_id,
                        'inning': inning,
                        'over': over,
                        'last_5_over_sum': last_5_over_sum,
                        'last_5_over_wickets': last_5_over_wickets,
                        'current_runs': cumulative_runs
                    }
                    results.append(result)

# Convert the results to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
results_df
```

```python
final = pd.merge(total_runs_by_inning, results_df, on=['match_id', 'inning'])
final
```

```python
final_merged_df = pd.merge(final, merged_df, on=['match_id', 'inning', 'over'])
final_merged_df
```

```python
final_extracted_df = final_merged_df.rename(columns={
    'match_id': 'mid',
    'date': 'date',
    'venue': 'venue',
    'batting_team': 'bat_team',
    'bowling_team': 'bowl_team',
    'batter': 'batsman',
    'bowler': 'bowler',
    'current_runs': 'runs',
    'is_wicket': 'wickets',
    'over': 'overs',
    'last_5_over_sum': 'runs_last_5',
    'last_5_over_wickets': 'wickets_last_5',
    'non_striker': 'non-striker',
    'total_runs_x': 'total'
}).filter(items=['mid', 'date', 'venue', 'bat_team', 'bowl_team', 'batsman', 'bowler', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5', 'non-striker', 'total'])
```

```python

final_extracted_df.to_csv("dataset.csv")
```
